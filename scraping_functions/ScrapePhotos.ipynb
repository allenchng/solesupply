```python
import pandas as pd
import urllib
import time
import requests
import random
```

```python
data = pd.read_csv('sneaker_metadata.csv')
```

```python
data_sub = data[~data.image.str.contains("--")]
```

```python
data_sub.head(10)
```

```python
for i in range(len(data_sub)):
    tmp = data_sub.iloc[i,:]
    
    output_dir = './sneaker_pictures/'
    
    sku = tmp['sku']
    url = tmp['image']
    file_path = output_dir + sku + '.jpg'
    
    with open(file_path, 'wb') as f:
        f.write(requests.get(url).content)
        
    time.sleep(random.randint(3,10))
```
